```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd.options.display.max_columns = None
```

```python
recensement_des_equipements_sportifs= pd.read_excel("./data_workshop/recensement-des-equipements-sportifs.xlsx")
```

### Analyse des instalations sportives

```python
installations_sportives = recensement_des_equipements_sportifs.loc[:, :"Installation hors-service"]
installations_sportives["Département"] = recensement_des_equipements_sportifs["Département"]
installations_sportives_sans_doublons = installations_sportives.drop_duplicates()
installations_sportives_sans_doublons.head()
```

```python
installations_sportives["Type de particularité de l'installation"].unique()
```

```python
installations_sportives_exploded = installations_sportives.assign(
    **{'Type de particularité de l\'installation': installations_sportives['Type de particularité de l\'installation'].str.split(',')}
).explode('Type de particularité de l\'installation').reset_index(drop=True)
installations_sportives_exploded["Type de particularité de l'installation"].unique()
```

```python
installations_sportives_sans_doublons["Accessibilité de l'installation en faveur des personnes en situation de handicap"] = installations_sportives_sans_doublons["Accessibilité de l'installation en faveur des personnes en situation de handicap"].astype(str)
installations_sportives_sans_doublons["Accessibilité de l'installation en faveur des personnes en situation de handicap"] = installations_sportives_sans_doublons["Accessibilité de l'installation en faveur des personnes en situation de handicap"].replace('nan', 'not_mentioned')
installations_sportives_sans_doublons["Accessibilité de l'installation en fonction du type handicap"] = installations_sportives_sans_doublons["Accessibilité de l'installation en fonction du type handicap"].fillna('aucun')
```

```python
installations_sportives_sans_doublons["Accessibilité de l'installation en fonction du type handicap"] = installations_sportives_sans_doublons["Accessibilité de l'installation en fonction du type handicap"].replace('Handicap moteur;Handicap sensoriel', 'Handicap moteur et sensoriel')
installations_sportives_sans_doublons["Accessibilité de l'installation en fonction du type handicap"] = installations_sportives_sans_doublons["Accessibilité de l'installation en fonction du type handicap"].replace('Handicap sensoriel;Handicap moteur', 'Handicap moteur et sensoriel')
```

```python
grouped_data_access = installations_sportives_sans_doublons.groupby("Accessibilité de l'installation en faveur des personnes en situation de handicap").size().reset_index(name='counts')
grouped_data_access
```

```python
plt.barh(grouped_data_access['Accessibilité de l\'installation en faveur des personnes en situation de handicap'], grouped_data_access['counts'])
plt.xlabel('Count')
plt.ylabel('Accessibilité de l\'installation')
plt.title('Accessibilité de l\'installation en faveur des personnes en situation de handicap')
plt.show()
```

```python
column_name = 'Département'
status_column = 'Accessibilité de l\'installation en faveur des personnes en situation de handicap'
 
pivot_table = pd.pivot_table(installations_sportives_sans_doublons, index=['Département'], columns=['Accessibilité de l\'installation en faveur des personnes en situation de handicap'], aggfunc='size', fill_value=0)
 
pivot_table_percentage = pivot_table.div(pivot_table.sum(axis=1), axis=0) * 100
 
fig, ax = plt.subplots(figsize=(12, 6))
pivot_table_percentage.plot(kind='bar', ax=ax)
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.text(x + width/2, y + height/2, f'{height:.1f}%', horizontalalignment='center', verticalalignment='center')
 
ax.set_title(f'Accessibilité de l\'installation en faveur des personnes en situation de handicap par {column_name}')
ax.set_xlabel(column_name)
ax.set_ylabel('Pourcentage')
ax.legend(title=status_column)

plt.show()
```

```python
column_name = 'Type d\'installation'
status_column = 'Accessibilité de l\'installation en faveur des personnes en situation de handicap'
 
pivot_table = pd.pivot_table(installations_sportives_exploded, index=["Type de particularité de l'installation"], columns=['Accessibilité de l\'installation en faveur des personnes en situation de handicap'], aggfunc='size', fill_value=0)
 
pivot_table_percentage = pivot_table.div(pivot_table.sum(axis=1), axis=0) * 100
 
fig, ax = plt.subplots(figsize=(12, 6))
pivot_table_percentage.plot(kind='bar', ax=ax)
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.text(x + width/2, y + height/2, f'{height:.1f}%', horizontalalignment='center', verticalalignment='center')
 
ax.set_title(f'Accessibilité de l\'installation en faveur des personnes en situation de handicap par {column_name}')
ax.set_xlabel(column_name)
ax.set_ylabel('Pourcentage')
ax.legend(title=status_column)

plt.show()
```

```python
grouped_data_access_air_jeux = recensement_des_equipements_sportifs.groupby("Accessibilité aux personnes à mobilité réduite à l'aire de jeu").size().reset_index(name='counts')
grouped_data_access_cheminement = recensement_des_equipements_sportifs.groupby("Accessibilité aux personnes à mobilité réduite aux cheminements").size().reset_index(name='counts')
grouped_data_access_tribunes = recensement_des_equipements_sportifs.groupby("Accessibilité aux personnes à mobilité réduite aux tribunes").size().reset_index(name='counts')
```

```python
grouped_data_access_air_jeux['Accessibilité aux personnes à mobilité réduite à l\'aire de jeu'] = grouped_data_access_air_jeux['Accessibilité aux personnes à mobilité réduite à l\'aire de jeu'].astype(str)
grouped_data_access_cheminement['Accessibilité aux personnes à mobilité réduite aux cheminements'] = grouped_data_access_cheminement['Accessibilité aux personnes à mobilité réduite aux cheminements'].astype(str)
grouped_data_access_tribunes['Accessibilité aux personnes à mobilité réduite aux tribunes'] = grouped_data_access_tribunes['Accessibilité aux personnes à mobilité réduite aux tribunes'].astype(str)
```

```python
grouped_data_access_cheminement
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

axes[0].barh(grouped_data_access_air_jeux["Accessibilité aux personnes à mobilité réduite à l'aire de jeu"], grouped_data_access_air_jeux['counts'])
axes[0].set_xlabel('Count')
axes[0].set_ylabel('Accessibilité aux personnes à mobilité réduite à l\'aire de jeu')
axes[0].set_title('Accessibilité aux personnes à mobilité réduite à l\'aire de jeu')

axes[1].barh(grouped_data_access_cheminement["Accessibilité aux personnes à mobilité réduite aux cheminements"], grouped_data_access_cheminement['counts'])
axes[1].set_xlabel('Count')
axes[1].set_ylabel('Accessibilité aux personnes à mobilité réduite aux cheminements')
axes[1].set_title('Accessibilité aux personnes à mobilité réduite aux cheminements')

axes[2].barh(grouped_data_access_tribunes["Accessibilité aux personnes à mobilité réduite aux tribunes"], grouped_data_access_tribunes['counts'])
axes[2].set_xlabel('Count')
axes[2].set_ylabel('Accessibilité aux personnes à mobilité réduite aux tribunes')
axes[2].set_title('Accessibilité aux personnes à mobilité réduite aux tribunes')

plt.tight_layout()
plt.show()
```
